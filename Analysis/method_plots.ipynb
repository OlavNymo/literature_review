## Screening Sankey diagram

```python
import plotly.graph_objects as go

labels = [
    "SCOPUS (792)",
    "IEEE Xplore (43)",
    "PROQUEST (489)",
    "WOS (825)",
    "Total (2149)",
    "Duplicates removed (946)",
    "Unique after dedup (1203)",
    "First screening included (114)",
    "First screening excluded (1089)",
    "Second screening included (50)",
    "Second screening excluded (64)",
]

# Index mapping for readability
S, I, P, W, C, D, U, F_in, F_out, S_in, S_out = range(len(labels))

sources = [S, I, P, W, C, C, U, U, F_in, F_in]
targets = [C, C, C, C, D, U, F_in, F_out, S_in, S_out]
values  = [792, 43, 489, 825, 946, 1203, 114, 1089, 50, 64]

fig = go.Figure(data=[go.Sankey(
    arrangement="fixed",
    node=dict(
        label=labels,
        pad=15,
        thickness=20,
        x=[0.0, 0.0, 0.0, 0.0, 0.15, 0.35, 0.35, 0.55, 0.55, 1, 0.7],
        y=[0.10, 0.30, 0.50, 0.70, 0.42, 0.75, 0.24, 0.03, 0.4, 0.03, 0.2],
        color="#8A8A8A",
        line=dict(color="#8A8A8A")
    ),
    link=dict(source=sources, target=targets, value=values, color="rgba(128,128,128,0.45)")
)])

fig.update_layout(font_size=12)
fig.show()
```
